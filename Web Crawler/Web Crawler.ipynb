```python
def news_crawler(web,newspaper_name,category):
    url_links = []
    urls_set = set()
    news_articles = newspaper.build(web, language = 'en', memoize_articles=False)
    for article in news_articles.articles:
       # check to see if the article url is not within the urls_set
       if article.url not in urls_set:
         # add the unique article url to the set
         urls_set.add(article.url)
         url_links.append(article.url)

    urls=url_links
    rows = []
    for url in urls:
        try:
            a = Article(url,language = "en")
            a.download()
            a.parse()
            # NLP on the article
            a.nlp()

            title = a.title
            article_title_lower = title.lower()
            text = a.text
            summary = a.summary
            keyword = a.keywords
            language = detect(summary)
            #print(language)

            if (summary != '' and language == 'en'and 'video' not in article_title_lower and newspaper_name not in article_title_lower):
                if(newspaper_name == "cnn"):
                    text=text.replace("(CNN) - ", '')
                    text=text.replace("CNN", '')
                elif(newspaper_name == "fox"):
                    text=text.replace("NEW You can now listen to Fox News articles!", '')
                    text=text.replace("Fox", '')
                    text=text.replace("CLICK HERE TO GET THE FOX NEWS APP", '')
                    text=text.replace("FOX", '')
                elif(newspaper_name == "abc"):
                    text=text.replace("ABC", '')
                    text=text.replace("contributed to this report", '')
                elif(newspaper_name == "nbc"):
                    text=text.replace("NBC", '')
                    text=text.replace("Twitter", '')
                data = {'newspaper':newspaper_name, 'category':category,'title':article_title, 'keywords':article_keywords,'summary':article_summary,'text':article_text}
                df = df.append(data,ignore_index=True)
                
                row = {'url':url,
                       'newspaper':newspaper_name,
                       'category':category,
                       'title':title,
                       'text':text,
                       'summary':summary,
                       'keyword':keyword}

                rows.append(row)
        except Exception as e:
            print(e)
            row = {'url':url,
            'newspaper':newspaper_name,
            'category':category,
            'title':'N/A',
            'text':'N/A',
            'summary':"N/A",
            'keyword':"N/A"}

            rows.append(row)

    df = pd.DataFrame(rows)   

    df.drop(df.index[df["text"]=="N/A"],inplace = True)
    df.drop(df.index[df["summary"]==""],inplace = True)
    df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
    df = df[df["url"].str.contains("video")==False]
    df = df.reset_index()
    df = df.drop('index', axis=1)
    return df
```

```python
from newspaper import build
from newspaper import Article
import newspaper

cnn_links = []
urls_set = set()
#cnn_articles = build('http://cnn.com', memoize_articles=False)
cnn_articles = build('https://www.cnn.com/politics', language = 'en', memoize_articles=False)
for article in cnn_articles.articles:
   # check to see if the article url is not within the urls_set
   if article.url not in urls_set:
     # add the unique article url to the set
     urls_set.add(article.url)
     cnn_links.append(article.url)


print(len(cnn_links))
```

```python
import pandas as pd
from langdetect import detect

urls=cnn_links
rows = []
for url in urls:
    try:
        a = Article(url,language = "en")
        a.download()
        a.parse()
        # NLP on the article
        a.nlp()
         
        title = a.title
        author = a.authors
        date = a.publish_date
        text = a.text
        summary = a.summary
        keyword = a.keywords
        language = detect(summary)
        print(language)
        
        if (summary != '' and language == 'en'):
            row = {'url':url,
                   'title':title,
                   'author':author,
                   'data':date,
                   'text':text,
                   'summary':summary,
                   'keyword':keyword}

            rows.append(row)
    except Exception as e:
        print(e)
        row = {'url':url,
        'title':'N/A',
        'author':'N/A',
        'data':'N/A',
        'text':'N/A',
        'summary':"N/A",
        'keyword':"N/A"}
        
        rows.append(row)
        
df_cnn = pd.DataFrame(rows)   
df_cnn
```

```python
df_cnn.drop(df_cnn.index[df_cnn["text"]=="N/A"],inplace = True)
df_cnn.drop(df_cnn.index[df_cnn["summary"]==""],inplace = True)
df_cnn.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df_cnn = df_cnn.reset_index()
df_cnn = df_cnn.drop('index', axis=1)
```
