```python
import numpy as np
import pandas as pd
import newspaper
from newspaper import build
from newspaper import Article
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0

#create empty dataframe
df = pd.DataFrame(columns = ['newspaper', 'category','title','keywords','summary'])

newspaper_name='cnn'
category='politics'

cnnnews=newspaper.build('https://www.cnn.com/politics',memoize_articles=False)
for article in cnnnews.articles:
    print(article.url)
    url=article.url
    
    #exclude error urls
    try:
        article_final=Article(url,language="en")
        article_final.download()

        article_final.parse()
        article_final.nlp()

        article_summary=article_final.summary

        #if summary is empty, it is video, so do not add to dataframe
        if article_final.summary != "" and detect(article_final.summary)== 'en':
            data = {'newspaper':newspaper_name, 'category':category,'text': article_final.text, 'keywords':article_final.keywords,'summary':article_final.summary}
            # print(data)
            df = df.append(data,ignore_index=True)
    except:
        continue
    
    
                
df.to_csv("cnn_politics.csv")
```
