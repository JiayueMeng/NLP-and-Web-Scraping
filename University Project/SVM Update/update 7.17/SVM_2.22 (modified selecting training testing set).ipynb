```python
#This is the result for testing between Physics, History, and Computer Science apartment of four colleges
#UR, UB, VAN, MIT, STAN
#The dataset has a total size of 752,
#We chose 20 words that matters the most to form the dataset, 
#and label data from History as "1", data from Physics as "2", and data from Computer Science as "3".
```

```python
#This is the part to do some basic process about the dataset. For the firt 163 rows, which is the data from history,
#add in a label "1". And for the following 285 rows, which is the data from Physics, add in a label "2".
#For the last 307 rows, which is the data from Computer Science, add in a label "3".
#Then, to merge the three datasets (History + Physics + Computer Science) to get the final dataset.
smalltest = pd.read_csv("data_ult1.csv") #Import the dataset.
smalltest = smalltest.drop('Unnamed: 0', axis = 1)
smalltest = smalltest.iloc[:, :20]
smalltest
```

```python
filt = smalltest.iloc[0:162] #Get the first 163 rows (data for History)
filt['output'] = [1 for row in range(len(filt))] #Add in a label "1" for the first 118 rows
filt2 = smalltest.iloc[162:446] #Get the next 285 rows (data for Physics)
filt2['output'] = [2 for row in range(len(filt2))] #Add in a label "2" for the next 270 rows
filt3 = smalltest.iloc[446:752] #Get the last 307 rows (data for Computer Science)
filt3['output'] = [3 for row in range(len(filt3))] #Add in a label "3" for the last 222 rows
smalltest = pd.merge(filt, filt2, how = 'outer') #Merge the three datasets for History, Physics, and Computer Science
smalltest = pd.merge(smalltest, filt3, how = 'outer')
smalltest
```

```python
#This is the part to separate training set and testing set. I would pick a size of 15 as testing set, which includes
#5 data from History, 5 data from Physics, and 5 data from Computer Science.
#To pick testing set, randomly select three locations in three categories and by adding 3 each time, get 5 different 
#locations in each category.
#Afterwards, the testing set would be used to see how many words would be accurately predicted.
#We would repeat this process for 100 times to get the average number of words that are correctly predicted.
```

```python
answer = []
count = 0
for i in range(100):
    smalltest_new = pd.merge(filt, filt2, how = 'outer') #Merge the three datasets for History, Physics, and Computer Science
    smalltest_new = pd.merge(smalltest_new, filt3, how = 'outer')
    loc_1 = random.randint(0, 160)
    loc_2 = random.randint(170, 440)
    loc_3 = random.randint(450,730)
    #print("loc_1: ", loc_1)
    #print("loc_2: ", loc_2)
    #print("loc_3: ", loc_3)

    #This part arims at separating testing and training set from the whole dataset.
    test_x = smalltest_new.iloc[[loc_1, loc_1+3, loc_1+6, loc_1+9, loc_1+12, loc_2, loc_2+3, loc_2+6, loc_2+9, loc_2+12, loc_3, loc_3+3, loc_3+6, loc_3+9, loc_3+12], :]
    train_x = smalltest_new.drop([loc_1, loc_1+3, loc_1+6, loc_1+9, loc_1+12, loc_2, loc_2+3, loc_2+6, loc_2+9, loc_2+12, loc_3, loc_3+3, loc_3+6, loc_3+9, loc_3+12])
    test_y = np.array(test_x['output'])
    test_x = test_x.drop('output', axis = 1)
    train_y = np.array(train_x['output'])
    train_x = train_x.drop('output', axis = 1)
    
    #This is the part for SVM. We choose the value of C and kernel function
    lin = svm.SVC(kernel = 'rbf', C = 100)
    lin.fit(train_x, train_y)
    predict = lin.predict(test_x)
    
    #This is the aprt of KNN. We choose the number of clusters and weights
    #knn = neighbors.KNeighborsRegressor(14, weights="uniform")
    #knn.fit(np.array(train_x), np.array(train_y))
    #predict = knn.predict(test_x)
    
    acc_num = sum(predict == test_y)
    answer.append(acc_num)

accuracy = np.average(answer)

print("accuracy:", accuracy)
    
```
